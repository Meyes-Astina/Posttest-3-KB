Untuk memlakukan visualisasi data harus menggunakan matplotlib dan seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
df.shape
```

## Bar Plot

```python
# pandas
# Buat bar plot untuk kolom "cut"
cut_counts = df['cut'].value_counts()
cut_counts.plot(kind='bar')
plt.title('Distribusi Jenis Pemotongan Berlian')
plt.xlabel('cut')
plt.ylabel('Jumlah')
plt.show()
```

Dari bar plot yang menggambarkan distribusi jenis pemotongan berlian, kita dapat membuat beberapa analisis:

Jenis pemotongan "Ideal" dan "Premium" adalah yang paling banyak dalam dataset. Ini menunjukkan bahwa jenis pemotongan ini adalah yang paling umum dalam data berlian yang diamati.
Jenis pemotongan "Fair" memiliki jumlah yang paling sedikit dalam dataset. Ini menunjukkan bahwa jenis pemotongan ini mungkin kurang diminati dalam data atau mungkin lebih jarang ditemui dalam pasar berlian.

## Scatter Plot

```python
# Membuat scatter plot untuk pengaruh "carat" terhadap "harga"
plt.hexbin(df['carat'], df['price'], gridsize=50, cmap='Blues')
plt.title('Pengaruh Carat terhadap Harga Berlian')
plt.xlabel('Carat')
plt.ylabel('Harga')
plt.colorbar(label='Jumlah')
plt.show()
```

Dapat terlihat bahwa ada hubungan positif antara berat berlian (carat) dan harga. Semakin besar berat berlian (semakin ke kanan), harga berlian cenderung lebih tinggi. Ini adalah temuan yang diharapkan, karena berat berlian adalah salah satu faktor utama yang memengaruhi harga.

## Pie Chart

```python
# Hitung jumlah setiap jenis pemotongan berlian
cut_counts = df['cut'].value_counts()

# Buat pie chart
plt.figure(figsize=(6, 6))
plt.pie(cut_counts, labels=cut_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Jenis Pemotongan Berlian')
plt.axis('equal')  # Memastikan pie chart menjadi lingkaran
plt.show()
```

Dalam diagram di atas, menampilkan jumlah setiap jenis pemotongan berlian dan pie chart yang menampilkan distribusi persentase jumlah berlian untuk setiap jenis pemotongan. Ini akan memberikan gambaran visual tentang seberapa banyak berlian yang termasuk dalam setiap kategori pemotongan (cut). 

Dapat disimpulkan bahwa berlian dengan jenis potongan Ideal adalah yang paling banyak yaitu 40.0% sedangkan jumlah jenis pemotongan Fair sangat yang paling sedikit yaitu 3.0%.

## Heatmap

```python
# Pilih hanya kolom-kolom numerik untuk heatmap korelasi
numeric_cols = ['carat', 'depth', 'table', 'price',]
numeric_df = df[numeric_cols]

# Membuat heatmap korelasi antara kolom-kolom numerik
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antara Kolom Numerik')
plt.show()
```

Dari heatmap korelasi ini, kita dapat menyimpulkan bahwa berat berlian (carat) adalah faktor utama yang memengaruhi harga, sementara kedalaman dan lebar bagian atas (depth dan table) memiliki korelasi yang lebih lemah dengan harga.

## Histogram

```python
# Pengelompokan data berdasarkan kolom "cut" dan menghitung statistik deskriptif
cut_groups = df.groupby('cut')['price']
cut_summary = cut_groups.describe()

# Visualisasi pengaruh harga dari "cut"
cut_summary['mean'].plot(kind='bar', color='skyblue', edgecolor='k')
plt.xlabel('Cut')
plt.ylabel('Rata-rata Harga')
plt.title('Pengaruh Cut terhadap Harga Berlian')
plt.show()
```

Dari histogram di atas dapat dilihat bahwa jenis pemotongan ("cut") memengaruhi harga berlian secara signifikan.
Jenis pemotongan "Ideal" dan "Premium" memiliki rata-rata harga berlian yang lebih tinggi dibandingkan dengan jenis pemotongan lainnya.
Ini menunjukkan bahwa dalam dataset ini, berlian dengan jenis pemotongan "Ideal" atau "Premium" cenderung lebih mahal.
Sedangkan jenis pemotongan "Good," "Very Good," dan "Fair" memiliki rata-rata harga yang lebih rendah dibandingkan dengan "Ideal" dan "Premium."
